## T1) Read the txt-file “Verne_Around_the_World.txt” into a string. The txt contains the book “Around the World in Eighty Days” (by Jules Verne; downloaded from gutenberg.org). (The book starts from chapter 1, and has been processed with split() and “ “.join() for removing line breaks and whitespaces.)

```python
with open("./Verne_Around_the_World.txt", mode="r", encoding="utf-8") as file:
    book = file.read()
```

## T2) Import spacy and the English language model. Use spacy nlp() to process the text into a doc object, which you can name “text_doc”. (You can only process first 50000 characters of the book to be a bit faster). 


```python
#importing spacy
import spacy

nlp = spacy.load("en_core_web_sm")
```

```python
text_doc = nlp(book[0:50000])
```

## T3) Create a list “persons_list = []”. Use a for-loop to iterate all the named entities (text_doc.ents), and store all the person entities (entity.label == “PERSON”) to the list. Print the persons_list. 


```python
persons_list = []

for entity in text_doc.ents:
    if entity.label_ == "PERSON":         
        persons_list.append(entity.text)


print(persons_list)
print("")

print("The entities tagged with label PRG are: " + spacy.explain("NORP"))
```

## T4) Use Counter.most_common() (from collections) to count the top five person names in your list of person names. 

```python
from collections import Counter

print ("All frequencies counted: " + str(Counter(persons_list)))
```
