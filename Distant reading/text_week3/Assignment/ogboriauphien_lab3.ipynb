## OPTIONAL T6)There is also a POS tag for proper nouns, “PROPN”, that has a lot in common with named entities. Pick one of the top person names in the novel (T4), and count all the cases where spacy has identified that name as a proper noun (“PROPN”). Discuss: when might POS tags be a better choice than using NE classes? For what kind of questions does only NER work? 

```python


# Get the person name
person_name = "Fogg"

# Initialize a count variable
count = 0

# Loop through each token in the processed text
for token in text_doc:
    # Check if the token is a proper noun and if its text is equal to the person name
    if token.pos_ == "PROPN" and token.text == person_name:
        # Increment the count
        count += 1

# Print the result
print("The name '{}' appeared {} times as a proper noun in the text.".format(person_name, count))
```

- Using POS tags like "PROPN" can be useful in cases when you want to count the frequency or occurences of a certain pronoun in a text corpus. NE, on the other hand is useful when you want to extract some specific named entities, for example, organisations, persons, or places. 
- POS tags, i believe explains what role a word plays in a sentence and enables the reader or analysizer have a better understanding of a term in a corpus. An example would be using the word "bank" can have different meanings depending on the context, such as a financial institution or the side of a river. POS tags can help determine which meaning is intended by providing information about the grammatical role of the word in a sentence.

## T14) Write a for-loop where you go through the speeches (at least 500 first speeches, it is a big file; try opening it with excel or notepad++ if you like!) and extract all organization/institution names (label_ == “ORG”).

```python
list_ORG = []

for speech in speeches[0:500]: 
    doc = nlp(speech, disable = ["lemmatizer", "tagger"])

    for entity in doc.ents:
        if entity.label_ == "ORG":
            list_ORG.append(entity.text) 
```
